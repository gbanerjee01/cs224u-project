```python
import json

with open('../CTC_baseline_splits/train.json') as json_file:
    ctc_train = json.load(json_file)
    
with open('../CTC_baseline_splits/dev.json') as json_file:
    ctc_dev = json.load(json_file)

with open('../gold_transcripts.json') as json_file:
    gold = json.load(json_file)  
```

```python
import pandas as pd

train_x = []
train_y = []

for key, val in ctc_train.items():
    train_x.append(val)
    train_y.append(gold[key])

df = pd.DataFrame(list(zip(train_x, train_y)), columns=["input_text", "target_text"])
```

```python
df
```

```python
df.to_pickle("train.pkl")
```
